```python
%load_ext autoreload

%matplotlib inline
%autoreload 2

# Supress nosiy deprecation warning from inside matplotlib
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import scipy.misc
```

```python
import otsu

def plot_inout(input_image, output_image):
    fig = plt.figure(figsize=(15, 15))
    
    input = fig.add_subplot(221)
    input.imshow(input_image)
    input_hist = fig.add_subplot(223)
    input_hist.hist(input_image.flatten(), bins=256, range=(0, 256), normed=True)
    
    output = fig.add_subplot(222)
    output.imshow(output_image, cmap='gray')
    output_hist = fig.add_subplot(224)
    output_hist.hist(output_image.flatten(), bins=256, range=(0, 256), normed=True)
    
    fig.show()

def test_example():
    p = 'bie_threshold.jpeg'
    img = scipy.misc.imread(p)
    binary = otsu.threshold(img, 150)
    plot_inout(img, binary)
    return None
    
test_example()
```
