```python
def read(p):
    columns = ('i', 'time', 'x', 'y', 'z')
    data = pandas.read_csv(p, header=None, skiprows=0, names=columns, index_col=(0))
    data.index = data['time'] / 1000
    return data

def plot_axes(df):
    df['x'].plot()
    df['y'].plot()
    df['z'].plot()

data = read('./data/lying-still.csv')
znorm = data['z'].mean()
znorm
```

```python

def est_altitude(data):
    zg = data['z'].rolling(5, center=False).mean() - znorm
    return zg.cumsum().cumsum()

scenarios = [
    'hovering',
    'lying-still',
    'table-to-hover',
    #'fast-up-down',
    'slowly-up-then-down',
    'slow-up-down',
    #'slowly-updown2',
]
data = { s: est_altitude(read('./data/'+s+'.csv')) for s in scenarios }

df = pandas.DataFrame(data)
df.plot()
```
